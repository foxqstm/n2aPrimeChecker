```python
from CheckPrime import is_prime
from ModulesFactorization import *
import matplotlib.pyplot as plt
import numpy as np

print("n^2-aが素数になる確率を計算します")

a_min_str=input("自然数aの最小値を入力してください:")
a_min=PositiveInt(a_min_str)

a_max_str=input("自然数aの最大値を入力してください:")
a_max=PositiveInt(a_max_str)

num_str=input("各aに対して調べる自然数の個数Numを入力してください:")
num=PositiveInt(num_str)
    
a_list=[]
Pa_list=[]
max_Pa=0
max_a=0

for a in range(a_min,a_max+1):
    if not is_square(a):
        ini_n=round(a**0.5)+1
        a_list.append(a)
        count_prime=0
        for n in range(ini_n,ini_n+num):
            p=n**2-a
            if is_prime(p):
                count_prime+=1
                #print(p)
        Pa=count_prime/num
        if Pa > max_Pa:
            max_Pa=Pa
            max_a=a
        Pa_list.append(Pa)
        #print(count_prime/num)

        
#print(a_list,Pa_list)
plt.plot(a_list,Pa_list,color="#77ff77",marker="o",linestyle=":") #エメラルドグリーンでプロット
plt.plot(max_a,max_Pa,color="red",marker="v",label="a={0},Pa={1}".format(max_a,max_Pa))

plt.xlabel("a")
plt.ylabel("P(a)")
plt.legend()

plt.savefig("n2a_a_min={0}_a_max={1}_Num={2}".format(a_min,a_max,num))

plt.show()
```
